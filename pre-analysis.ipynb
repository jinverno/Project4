```python
data.dropna(inplace=True)
data.drop(axis=0,labels=data[data["name"]=="Null"].index,inplace=True)
data.info()
```

```python
x=data[data["name"].str.lower().str.contains("wallet")].groupby(by="category").mean()
x["ratio"]=x["usd_pledged_real"]/x["usd_goal_real"]
x.sort_values(by="ratio", ascending=False)
```

```python
y=data[data["name"].str.lower().str.contains("djin")]
y["ratio"]=y["usd_pledged_real"]/y["usd_goal_real"]
y.sort_values(by="ratio", ascending=False)
```
